MRの定員充足率(指定科類枠)

```python
trial=1000
fill_dict=dict()
for f in fac:
    temp=dict()
    for key in set(minorityReserveDA[f].keys())-set(["0", "123456"]):
        temp[key]=0
    fill_dict[f]=temp
for n in range(trial):
    sim_fill=np.load(f"MultiThread/sim_fill{n}.npy", allow_pickle=True).item()
    for f in fac:
        mr_dict=sim_fill[f]
        for key in set(mr_dict.keys())-set(["0", "123456"]):
            mr_sim=len(mr_dict[key])
            mr_row=minorityReserveDA[f][key]
            if mr_row != 0:
                fill_dict[f][key]+=mr_sim/mr_row
            else:
                fill_dict[f][key]+=0
for f in fac:
    for key in fill_dict[f]:
        fill_dict[f][key]=fill_dict[f][key]/trial
print(fill_dict)
```

```python
trial=1000
amount_of_acceptance_MQ=dict()
count_0_MQ=dict()
for f in fac:
    amount_of_acceptance_MQ[f]=0
    count_0_MQ[f]=0
for n in range(trial):
    minority_acceptance_MQ=pd.read_csv(f"MultiThread/minority_acceptance_MQ{n}.csv", index_col=0, header=None)
    majority_acceptance_MQ=pd.read_csv(f"MultiThread/majority_acceptance_MQ{n}.csv", index_col=0, header=None)
    for f in fac:
        amount_of_acceptance_MQ[f]+=len(minority_acceptance_MQ.loc[f].dropna(0))
        amount_of_acceptance_MQ[f]+=len(majority_acceptance_MQ.loc[f].dropna(0))
        if len(majority_acceptance_MQ.loc[f].dropna(0))+len(minority_acceptance_MQ.loc[f].dropna(0)) == 0:
            count_0_MQ[f]+=1
        
for f in fac:
    amount_of_acceptance_MQ[f]=amount_of_acceptance_MQ[f]/trial
    count_0_MQ[f]={"受入数0の回数":count_0_MQ[f]}
for f in fac:
    amount_of_acceptance_MQ[f]={"受入数":amount_of_acceptance_MQ[f]}
df_aa_MQ=pd.DataFrame(amount_of_acceptance_MQ).T
df_count_0_MQ=pd.DataFrame(count_0_MQ).T
print(df_aa_MQ)
print(df_count_0_MQ)
df_aa_MQ.to_csv("Result/受入数MQ.csv", encoding="utf-8-sig")
df_count_0_MQ.to_csv("Result/受け入れ数0の回数_MQ.csv", encoding="utf-8-sig")
```
