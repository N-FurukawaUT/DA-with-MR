```python
import numpy as np
n=0
trial=1000
shitei_dict_sim=dict()
zenkarui_dict_sim=dict()
for f in fac:
    shitei_dict_sim[f]={"l1":0,"l2":0,"l3":0,"s1":0,"s2":0,"s3":0}    #batchの更新に使う
    zenkarui_dict_sim[f]={"l1":0,"l2":0,"l3":0,"s1":0,"s2":0,"s3":0}    #batchの更新に使う
for n in range(trial):
    sim_fill=np.load(f"MultiThread/sim_fill{n}.npy", allow_pickle=True).item()
    zenkarui_sim_df=pd.read_csv(f"MultiThread/majority_acceptance_MQ{n}.csv", header=None, index_col=0)
    df=pd.read_csv(f"MultiThread/created_submit_komaba{n}.csv", header=0, index_col=0)
    df=df.loc[:,"karui"]
    karui_dict=df.to_dict()
    for f in fac:
        karui_keys=list(sim_fill[f].keys())
        for k in karui_keys:
            accept_student=sim_fill[f][k]
            for student in accept_student:
                karui=karui_dict[student]
                shitei_dict_sim[f][karui]+=1
    for f in fac:
        accept_student=zenkarui_sim_df.loc[f].dropna().to_list()
        for student in accept_student:
            karui=karui_dict[student]
            zenkarui_dict_sim[f][karui]+=1
            
#print(shitei_dict_sim)
for f in fac:
    for k in ["l1","l2","l3","s1","s2","s3"]:
        zenkarui_dict_sim[f][k]=zenkarui_dict_sim[f][k]/trial
        shitei_dict_sim[f][k]=shitei_dict_sim[f][k]/trial

shitei_df=pd.DataFrame.from_dict(shitei_dict_sim, orient="index")
zenkarui_df=pd.DataFrame.from_dict(zenkarui_dict_sim, orient="index")

pd.set_option('display.max_rows', 100)
print(shitei_df)
print(zenkarui_df)
```

```python
n=0
trial=1000
utility_dict=dict()
kaisuu=dict()
for f in fac:
    target = minorityReserveDA[f]
    temp_dict=dict()
    for key in list(set(target.keys())-set(["0","123456"])):
        temp_dict[key]=0
    kaisuu[f]=temp_dict
    
    
for f in fac:
    temp_dict=dict()
    for key in list(set(minorityReserveDA[f].keys())-set(["0","123456"])):
        temp_dict[key]=0
    utility_dict[f]=temp_dict
for n in range(trial):
    student_dict=np.load(f"MultiThread/sim_fill{n}.npy", allow_pickle=True).item()
    utility=pd.read_csv(f"MultiThread/utilityDict{n}.csv", header=0, index_col=0)
    utility=utility.to_dict(orient="index")
    for f in fac:
        temp_dict=dict()
        target=student_dict[f]
        for key in list(set(target.keys())-set(["0","123456"])):
            worst_student_utility=10000
            for student in target[key]:
                student_utility=utility[student]["utility"]
                if student_utility < worst_student_utility:
                    worst_student_utility=student_utility
            if worst_student_utility!=10000:
                temp_dict[key]=worst_student_utility
                kaisuu[f][key]+=1
            else:
                temp_dict[key]=0
        for key in list(set(target.keys())-set(["0","123456"])):
            utility_dict[f][key]+=temp_dict[key]
for f in fac:
    target = utility_dict[f]
    for key in list(set(minorityReserveDA[f].keys())-set(["0","123456"])):
        if kaisuu[f][key]!=0:
            utility_dict[f][key]=utility_dict[f][key]/kaisuu[f][key]
        else:
            utility_dict[f][key]=0
print(utility_dict)
```
